```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Normalize the MFCCs (Min-Max Scaling)
mfcc_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))

# Compute Delta (Δ) and Delta-Delta (ΔΔ) features
delta_1 = librosa.feature.delta(mfcc_normalized)
delta_2 = librosa.feature.delta(mfcc_normalized, order=2)

# Combine static, delta, and delta-delta features
combined_features = np.vstack([mfcc_normalized, delta_1, delta_2])

# Compute the mean of each feature (across time axis, axis=1)
mean_features = np.mean(combined_features, axis=1)

# Save the mean features to CSV file
csv_file_path = 'mean_mfcc_delta_features.csv'
# df_mean = pd.DataFrame(mean_features)
df_mean = pd.DataFrame([mean_features], columns=mean_features)
df_mean.to_csv(csv_file_path, index=False, header=False)
print(f"Mean of MFCC, Δ, and ΔΔ features saved to {csv_file_path}")

# Visualize the features (Optional)
plt.figure(figsize=(10, 8))

# Static MFCCs
plt.subplot(3, 1, 1)
librosa.display.specshow(mfcc_normalized, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar()
plt.title('Normalized MFCCs')

# Delta (Δ)
plt.subplot(3, 1, 2)
librosa.display.specshow(delta_1, x_axis='time', sr=sr, cmap='plasma')
plt.colorbar()
plt.title('Delta (Δ) Features')

# Delta-Delta (ΔΔ)
plt.subplot(3, 1, 3)
librosa.display.specshow(delta_2, x_axis='time', sr=sr, cmap='inferno')
plt.colorbar()
plt.title('Delta-Delta (ΔΔ) Features')

plt.tight_layout()
plt.show()
```

```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Normalize the MFCCs (Min-Max Scaling)
mfcc_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))

# Compute Delta (Δ) and Delta-Delta (ΔΔ) features
delta_1 = librosa.feature.delta(mfcc_normalized)
delta_2 = librosa.feature.delta(mfcc_normalized, order=2)

# Combine static, delta, and delta-delta features
combined_features = np.vstack([mfcc_normalized, delta_1, delta_2])

# Compute the mean of each feature (across time axis, axis=1)
mean_features = np.mean(combined_features, axis=1)

# Compute an additional numerical value (e.g., overall mean of all features)
# overall_mean = np.mean(mean_features)
overall_mean = 999

# Append the single value to the mean features as the last column
mean_features_with_overall = np.append(mean_features, overall_mean)

# Save the updated mean features to CSV file
csv_file_path = 'mean_mfcc_delta_features_with_overall_mean.csv'
df_mean = pd.DataFrame([mean_features_with_overall])
df_mean.to_csv(csv_file_path, index=False, header=False)

print(f"Mean of MFCC, Δ, and ΔΔ features (with overall mean) saved to {csv_file_path}")
```
