```python
#  Extract and visualize default mfcc features
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features (default: 20 coefficients)
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Visualize the MFCC features
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfccs, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar(format="%+2.0f dB")
plt.title('MFCC (Mel Frequency Cepstral Coefficients)')
plt.tight_layout()
plt.show()
```

```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Normalize the MFCCs (Min-Max Scaling)
mfcc_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))

# Save normalized MFCC features to a CSV file
csv_file_path = 'normalized_mfcc_features.csv'
mfcc_df = pd.DataFrame(mfcc_normalized)
mfcc_df.to_csv(csv_file_path, index=False, header=False)
print(f"Normalized MFCC features saved to {csv_file_path}")

# Visualize the normalized MFCC features
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfcc_normalized, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar(format="%+2.0f")
plt.title('Normalized MFCC (Min-Max Scaling)')
plt.tight_layout()
plt.show()
```

```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Normalize the MFCCs (Min-Max Scaling)
mfcc_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))

# Compute Delta () and Delta-Delta () features
delta_1 = librosa.feature.delta(mfcc_normalized)
delta_2 = librosa.feature.delta(mfcc_normalized, order=2)

# Combine static, delta, and delta-delta features
combined_features = np.vstack([mfcc_normalized, delta_1, delta_2])

# Save to CSV file
csv_file_path = 'normalized_mfcc_delta_features.csv'
df = pd.DataFrame(combined_features)
df.to_csv(csv_file_path, index=False, header=False)
print(f"Normalized MFCC, , and  features saved to {csv_file_path}")

# Visualize the features
plt.figure(figsize=(10, 8))

# Static MFCCs
plt.subplot(3, 1, 1)
librosa.display.specshow(mfcc_normalized, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar()
plt.title('Normalized MFCCs')

# Delta ()
plt.subplot(3, 1, 2)
librosa.display.specshow(delta_1, x_axis='time', sr=sr, cmap='plasma')
plt.colorbar()
plt.title('Delta () Features')

# Delta-Delta ()
plt.subplot(3, 1, 3)
librosa.display.specshow(delta_2, x_axis='time', sr=sr, cmap='inferno')
plt.colorbar()
plt.title('Delta-Delta () Features')

plt.tight_layout()
plt.show()
```

```python
num_rows, num_columns = delta_1.shape
# Print the number of rows and columns
print("MCF")
print(f"Number of rows (MFCC d1 coefficients): {num_rows}")
print(f"Number of columns (Time d1 frames): {num_columns}")
```

```python
num_rows, num_columns = delta_2.shape
# Print the number of rows and columns
print("MCF")
print(f"Number of rows (MFCC d2 coefficients): {num_rows}")
print(f"Number of columns (Time d2 frames): {num_columns}")
```

```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'audio.wav'
y, sr = librosa.load(file_path)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Normalize the MFCCs (Min-Max Scaling)
mfcc_normalized = (mfccs - np.min(mfccs)) / (np.max(mfccs) - np.min(mfccs))

# Compute Delta () and Delta-Delta () features
delta_1 = librosa.feature.delta(mfcc_normalized)
delta_2 = librosa.feature.delta(mfcc_normalized, order=2)

# Combine static, delta, and delta-delta features
combined_features = np.vstack([mfcc_normalized, delta_1, delta_2])

# Compute the mean of each feature (across time axis, axis=1)
mean_features = np.mean(combined_features, axis=1)

# Save the mean features to CSV file
csv_file_path = 'mean_mfcc_delta_features.csv'
# df_mean = pd.DataFrame(mean_features)
df_mean = pd.DataFrame([mean_features], columns=mean_features)
df_mean.to_csv(csv_file_path, index=False, header=False)
print(f"Mean of MFCC, , and  features saved to {csv_file_path}")

# Visualize the features (Optional)
plt.figure(figsize=(10, 8))

# Static MFCCs
plt.subplot(3, 1, 1)
librosa.display.specshow(mfcc_normalized, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar()
plt.title('Normalized MFCCs')

# Delta ()
plt.subplot(3, 1, 2)
librosa.display.specshow(delta_1, x_axis='time', sr=sr, cmap='plasma')
plt.colorbar()
plt.title('Delta () Features')

# Delta-Delta ()
plt.subplot(3, 1, 3)
librosa.display.specshow(delta_2, x_axis='time', sr=sr, cmap='inferno')
plt.colorbar()
plt.title('Delta-Delta () Features')

plt.tight_layout()
plt.show()
```
